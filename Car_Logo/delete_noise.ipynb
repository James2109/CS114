```python
import os
import re
import pandas as pd

# List các tên loại xe hơi
df = pd.read_excel("Car.xlsx")
car_list = df["car_name"]

# Xóa các tệp tin logo_2, logo_3, logo_4 trong mỗi loại xe hơi trong list
for car in car_list:
    folder_name = car.replace(" ", "_")  # Chuyển tên xe hơi thành tên thư mục hợp lệ

    # Lấy danh sách tệp tin trong thư mục
    file_list = os.listdir(folder_name)

    # Xóa các tệp tin logo_2, logo_3, logo_4
    for file_name in file_list:
        match = re.match(r"logo_[1-4]\.jpg", file_name)
        if match:
            file_path = os.path.join(folder_name, file_name)
            os.remove(file_path)
            print(f"Đã xóa tệp tin {file_name} trong thư mục {folder_name}")
```
