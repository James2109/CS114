```python
import os
import time
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd

def crawl_logo_images(car_name, num_images, folder_name):
    # Tạo thư mục để lưu trữ ảnh logo
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)

    # Khởi tạo trình duyệt Chrome
    driver = webdriver.Chrome()

    try:
        # Tìm kiếm trên Google Images
        search_url = f"https://www.google.com/search?tbm=isch&q={car_name} logo"

        # Mở trang web
        driver.get(search_url)
        time.sleep(2)  # Đợi để trang web tải đầy đủ

        # Cuộn xuống để tải thêm ảnh
        last_height = driver.execute_script("return document.body.scrollHeight")
        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height

        # Lấy HTML của trang web sau khi đã cuộn xuống
        page_source = driver.page_source

        # Phân tích cú pháp HTML
        soup = BeautifulSoup(page_source, 'html.parser')

        # Tìm các thẻ <img> trong trang web
        img_tags = soup.find_all('img')

        # Lưu ảnh xuống ổ đĩa
        count = 0
        for img in img_tags:
            img_url = img.get('src')

            # Kiểm tra URL hợp lệ
            if img_url and img_url.startswith("http"):
                response = requests.get(img_url)
                if response.status_code == 200:
                    with open(f"{folder_name}/logo_{count+1}.jpg", "wb") as f:
                        f.write(response.content)
                        print(f"Lưu ảnh {count+1}/{num_images} của {car_name} thành công!")
                        count += 1

            # Đạt đủ số lượng ảnh
            if count >= num_images:
                break
    finally:
        # Đóng trình duyệt
        driver.quit()

# List các tên loại xe hơi
df = pd.read_excel("Car.xlsx")
car_list = df["car_name"]

# Crawl 100 ảnh có chứa logo cho mỗi loại xe hơi trong list
for car in car_list:
    folder_name = car  # Chuyển tên xe hơi thành tên thư mục hợp lệ
    crawl_logo_images(car, 104, folder_name)
```
